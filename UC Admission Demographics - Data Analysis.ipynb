# Abstract

The University of California admits thousands of students annually, however, the impact of  local demographics in this process remains unclear. This study analyzes demographic data regarding admitted students and California counties from 2021-2023 to determine whether there is a correlation between the two. Our findings reveal that there is minimal correlation between county-level demographic changes and UC admit demographics, indicating that other factors—such as holistic review, socioeconomic factors, and policy changes—may be at play in shaping UC admissions.

# Research Question

How do year-over-year changes in the demographic composition of UC System admits from 2021 to 2023 correlate with shifts in county-level population demographics over the same period, and what patterns emerge when comparing these trends across different regions in California?

## Background and Prior Work

The University of California (UC) system is one of the largest and most well-known public university systems in the country, admitting thousands of students from diverse backgrounds each year. The mix of students admitted to UC schools changes over time due to a variety of factors, like updates to admission policies, shifts in who is applying, and overall population trends in California. Understanding these patterns can help identify whether students from different regions are gaining admission at rates that reflect their local demographics and how well the UC system is adapting to California’s changing population. 

There has been a lot of significant policy changes lately. Some examples of this would be removing standardized testing requirements, expanding holistic admissions, and pushing towards improving equity and access. Simultaneously, California has experienced broader demographic shifts. Some of these shifts include declining birth rates, population movement between regions, socioeconomic changes. This study explores how county-level demographic trends from 2021 to 2023 relate to changes in UC admissions and whether distinct regional patterns emerge. 

Prior research has looked at how demographic factors influence college admissions.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The 2017 study by Reardon et al. found that racial and socioeconomic disparities in college enrollment are often associated with regional differences in economic conditions. A study by Bleemer (2022) looked at how affirmative action policies in California have shaped UC admissions.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) This directly translates to showing that changes in these policies directly impact the diversity of students being admitted. In relation to our research, it explores how shifts in county populations over time compare to changes in who gets admitted to UC schools each year. 

By looking into county-level population trends in addition to UC admit data, this study aims to find whether there are certain regions where admission trends mirror demographic shifts.

## Regression Analysis and Hypothesis Testing

```python
import statsmodels.formula.api as smf
```

To determine whether a linear relationship exists between the percent change in a race’s county population and the corresponding percent change in UC admits, we conduct a hypothesis test. But first, we perform a regression analysis to inform our decision-making for the following hypotheses:

For the hypothesis tests, we use the p-value as the test statistic, following the standard significance level of 0.05.

<ins>Null Hypothesis</ins> (${H_0}$): The change in the African American demographic of admitted UC students in each county is not affected by the change in the African American population in each county over the same period.

<ins>Alternate Hypothesis</ins> (${H_1}$): The change in the African American demographic of admitted UC students in each county is affected by the change in the African American population in each county over the same period.

```python
# rename columns for easier accessing
census_combined = census_combined.rename(columns={'African American_change': 'African_American_change'})
schools_combined = schools_combined.rename(columns={'African American_change': 'African_American_change'})

# merge census and schools data on county
merged_data = pd.merge(
    census_combined[['County', 'African_American_change', 'Asian_change', 'White_change']],
    schools_combined[['County', 'African_American_change', 'Asian_change', 'White_change']],
    on='County', how='inner', suffixes=('_census', '_schools')
)

# drop rows with any missing values in independent/dependent variables
merged_data_cleaned = merged_data.dropna(subset=['African_American_change_census', 'African_American_change_schools',
                                                 'Asian_change_census', 'Asian_change_schools', 
                                                 'White_change_census', 'White_change_schools'])
```

```python
# OLS Regression for African American demographic
model_aa = smf.ols(formula='African_American_change_schools ~ African_American_change_census', data=merged_data_cleaned)
res_aa = model_aa.fit()
print("REGRESSION FOR AFRICAN AMERICAN DEMOGRAPHIC:")
display(res_aa.summary())
pvalue_aa = res_aa.pvalues['African_American_change_census']
```

The p-value obtained is 0.323, which is much greater than the specified significance level of 0.05. # Ethics & Privacy

Our dataset mainly includes high schools in California, which means it does not account for out-of-state applicants or students from private schools, who may apply to UC schools at different rates. As a result of this, our final analysis might not accurately identify all the demographic trends in UC admissions. Additionally, the county demographic data we are using comes from census estimates, which heavily rely on self-reported survey responses. This could occasionally lead to misclassification of racial and ethnic identities, potentially affecting the accuracy of our findings. Another factor to consider is that population estimates come with margins of error. This can vary by county. Smaller counties or underrepresented populations may have greater uncertainty in their data. This makes direct comparisons even more difficult. To address these challenges, we will clearly acknowledge any limitations in our dataset when presenting our findings. When possible, we will adjust for margin-of-error ranges in demographic estimates and look into additional data sources, such as UC system-wide reports, to give a more complete picture of admission trends.

While our datasets don’t include any personally-identifiable information (PII), there is still a chance that students from smaller applicant pools (like those from rural schools or underrepresented communities) could be indirectly identified if our data were to be combined with other public resources. Since our analysis looks at racial and ethnic demographics, we also need to be mindful of how we present our findings to avoid reinforcing stereotypes or making assumptions about data that might be misleading. To minimize these risks, we can focus on county-level trends instead of individual schools. We will also be careful in how we interpret the data, sticking to observations instead of drawing conclusions about the impact of UC admission policies. 

Lastly, as our project looks at the relationships between California Counties demographic shifts and UC admissions, we face the risk of insinuating erasure of the many other external factors, such as income and school funding, that affect the admit rates of different schools to the UC System. Additionally, as the University of California participates in holistic admissions and we do not look at the underlying trends of personal statements and applications, we risk oversimplifying the changes within the admissions process to only demographic shifts. To minimize these challenges, we will conduct statistical analyses that control for those external confounders, to only focus on the planned relationship. This way, we only look at correlation and never imply causation.

To answer our research question, we need census datasets and University of California applicant information. In order to ethically use these datasets, the privacy must be analyzed. Both census data and UC applicant information are publically available, so using them does not violate any sort of privacy restrictions.

Additionally, in order to make the question ethically perceived, we must highlight that local demographics is only one factor that goes into the admissions process, and not the only factor: it does not overpower merit-based evaluations. We have handled this issue through explaining in our abstract that local demographics is not a significant enough factor to attribute admission decisions to.
